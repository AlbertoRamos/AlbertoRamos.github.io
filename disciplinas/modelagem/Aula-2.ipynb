# Programação Inteira

O problema de programação inteira é essencialmente o mesmo do de PL, com
a diferença notável de que as variáveis devem ser inteiras.
Caso apenas algumas sejam inteiras, podemos dizer programação inteira mista.

$$ \min c^Tx \\
\mbox{suj. a } Ax = b \\
x \geq 0 \\
x_i \in \mathbb{Z}, i \in I
$$

Em alguns casos, a solução do problema sem a restrição de integralidade
(dito problema relaxado) fornece a solução do problema original.
Em outros casos, a solução do problema relaxado arredondada posteriormente
também pode trazer a solução original.
No entanto, a solução pode, e muitas vezes é, bastante diferente dessa
aproximação relaxada.

Por esse motivo, a resolução de um problema misto é muito mais complicada.
Um problema com algumas dúzias de variáveis pode levar muito mais tempo para
ser resolvido que um problema linear contínua com milhares de variáveis.

Ex.:
$$ \max c_1x_1 + c_2x_2 \\
\mbox{suj. a }
0.1x_1 + x_2 \leq 3 \\
x_1 + 0.1x_2 \leq 2.9 \\
x_1, x_2 \geq 0.
$$

```julia
using JuMP, Cbc, Clp, Interact

@manipulate for θ = linspace(0, pi/2, 400)
    c = [cos(θ); sin(θ)]

    t = linspace(-1, 4, 100)
    contour(t, t, (x,y)->x*c[1]+y*c[2], levels=50)
    plot!(0*t, t, c = :blue, leg=false)
    plot!(t, 0*t, c = :blue)
    plot!(t, -0.1t + 3, c=:blue)
    plot!(-0.1t + 2.9, t, c=:blue)

    m = Model()
    @variable(m, x[1:2] >= 0)
    @objective(m, Max, c[1]*x[1] + c[2]*x[2])
    @constraint(m, 0.1*x[1]+x[2] <= 3)
    @constraint(m, x[1]+0.1*x[2] <= 2.9)
    solve(m)
    xlin = getvalue(x)
    m = Model()
    @variable(m, x[1:2] >= 0, Int)
    @objective(m, Max, c[1]*x[1] + c[2]*x[2])
    @constraint(m, 0.1*x[1]+x[2] <= 3)
    @constraint(m, x[1]+0.1*x[2] <= 2.9)
    solve(m)
    xint = getvalue(x)

    scatter!([xlin[1];xint[1]], [xlin[2];xint[2]], c=:red)

    xlims!(-0.5, 3.5)
    ylims!(-0.5, 3.5)
end
```

## TSP

O problema do caixeiro viajante (Travelling Salesman Problem) é um dos problemas clássicos
de PI.
O problema consiste em visitar uma quantidade finita de cidades, retornando para a cidade de
partida, visitando cada cidade uma única vez e buscando o menor caminho/custo de viagem.

```julia
N = 7
pontos = rand(2,N)
```

```julia
p = scatter(pontos[1,:], pontos[2,:], c=:blue, leg=false)
```

Seja $d_{i,j}$ é a distância (custo) da cidade $i$ até a cidade $j$, $1 \leq i,j \leq n$).
Denotamos por $x_{i,j}$, $1 \leq i,j \leq n$ a variável que indica se o caixeiro sai da
cidade $i$ para a cidade $j$.


$$ x_{i,j} = \left\{\begin{array}{ll}
    1, & \mbox{ se o mercador para sai de } i \mbox{ para } j, \\
    0, & \mbox{ caso contrário.}
    \end{array}\right. $$

Sendo assim, o custo da viajem do caixeiro é $\displaystyle\sum_{i=1}^n\sum_{j=1}^n d_{i,j}c_{i,j}$.

A seguir, devemos ter as restrições do caminho. Como passamos apenas uma vez em cada cidade,
pra cada cidade, nós chegamos nela apenas uma vez e saímos dela apenas uma vez.
Sendo assim, para um dado $i$, só existe um $j$ tal que $x_{i,j} = 1$,
e para um dado $j$, só existe um $i$ tal que $x_{i,j} = 1$. Isso se resume a
$$
\sum_{j=1}^nx_{i,j} = 1 \\
\sum_{i=1}^nx_{i,j} = 1.
$$

Até aqui tudo bem, no entanto isso não é suficiente para montar uma solução factível.

```julia
function plot_path(pontos, X)
    p = scatter(pontos[1,:], pontos[2,:], c=:blue, leg=false)
    N = size(pontos, 2)
    for i = 1:N
        for j = 1:N
            if X[i,j] == 1
                I = [i;j]
                plot!(p, pontos[1,I][:], pontos[2,I][:], c=:red)
            end
        end
    end
    return p
end

function ItoX(I)
    n = length(I)
    X = zeros(Int, n, n)
    for k = 1:n-1
        (j,i) = I[k:k+1]
        X[i,j] = 1
    end
    (j,i) = I[[n;1]]
    X[i,j] = 1
    return X
end
```

```julia
N = 10
t = linspace(0, 2pi, N)
pontos = [sin(t)'+t'; sin(t)']
X = ItoX([3;5;1;2;6;7;8;9;10;4])
plot_path(pontos, X)
```

```julia
N = 10
t = linspace(0, 2pi, N)
pontos = [sin(t)'+t'; sin(t)']
X = [ItoX([1;2;3;4;5]) zeros(5,5); zeros(5,5) ItoX([2;4;1;3;5])]
plot_path(pontos, X)
```

Precisamos evitar ciclos. Uma maneira de evitar isso é adicionar
as restrições
$$ \sum_{i \in K}\sum_{j \in K, j\neq i} x_{i,j} \leq \left\vert K\right\vert - 1,
\qquad 
\forall K \subsetneq \{1,2,\dots,N\}
$$

A negação desta condição é que existe um conjunto $K$, que não são todas as cidades,
tal que essa soma é $|K|$, isto é, existem $|K|$ viagens entre essas $K$ cidades,
e pelas outras condições, elas não se repetem. Portanto $K$ define um ciclo.

O "leve" problema decorrente disso é que o número de restrições é combinatorial
no número de cidades, o que é inviável.

Existem duas estratégias principais aqui.
Uma delas é a de resolver o problema e adicionar restrições a posteriori.

```julia
m = Model()
@variable(m, x[1:N,1:N], Bin)
@objective(m, Min, sum{norm(pontos[:,i]-pontos[:,j])*x[i,j], i=1:N, j=1:N})
@constraint(m, arrival[i=1:N], sum{x[i,j], j=1:N; i!=j} == 1)
@constraint(m, departures[j=1:N], sum{x[i,j], i=1:N; i!=j} == 1)
@constraint(m, semloop[i=1:N], x[i,i] == 0)
solve(m)
```

```julia
X = getvalue(x)
p = plot_path(pontos, X)
title!(p, "Distância total: $(getobjectivevalue(m))")
p
```

```julia
function temloop(X)
    n = length(X[:,1])
    visited = fill(false, n)
    i = 1
    while !visited[i]
        visited[i] = true
        i = findfirst(X[:,i] .== 1)
    end
    return !all(visited), find(visited)
end
```

```julia
temloop(X)
```

```julia
m = Model()
@variable(m, x[1:N,1:N], Bin)
@objective(m, Min, sum{norm(pontos[:,i]-pontos[:,j])*x[i,j], i=1:N, j=1:N})
@constraint(m, arrival[i=1:N], sum{x[i,j], j=1:N; i!=j} == 1)
@constraint(m, departures[j=1:N], sum{x[i,j], i=1:N; i!=j} == 1)
@constraint(m, semloop[i=1:N], x[i,i] == 0)

loop = true
iter = 0
while loop
    solve(m)
    # Checando se tem loop
    X = getvalue(x)
    loop, K = temloop(X)
    # Plotando
    p = plot_path(pontos, X)
    title!(p, "Distância total: $(getobjectivevalue(m))")
    png("tsp-intermediario-$iter")
    println(K)
    if loop
        @constraint(m, sum{x[i,j], i=K,j=K;i!=j} <= length(K)-1)
    end
    iter += 1
    if iter >= 100
        break
    end
end
p = plot_path(pontos, X)
title!(p, "Distância total: $(getobjectivevalue(m))")
p
```

A segunda é a usar variáveis extras indicando o caminho e usá-las
para limitar os movimentos.

Adicionamos
$$ 2 \leq u_i \leq n, \qquad i = 2,\dots,n \\
u_i - u_j + nx_{i,j} \leq n-1 \qquad 2 \leq i,j \leq n, i \neq j
$$

Note que, de fato, isso impede o ciclo, e que, ainda, $u_i$ acaba marcando
a posição da cidade $i$ no percurso, começando da cidade 1.

Se $x_{i,j} = 1$, então $u_j \geq u_i + 1$.

```julia
m = Model()
@variable(m, x[1:N,1:N], Bin)
@variable(m, 2 <= u[2:N] <= N)
@objective(m, Min, sum{norm(pontos[:,i]-pontos[:,j])*x[i,j], i=1:N, j=1:N})
@constraint(m, arrival[i=1:N], sum{x[i,j], j=1:N; i!=j} == 1)
@constraint(m, departures[j=1:N], sum{x[i,j], i=1:N; i!=j} == 1)
@constraint(m, semloop[i=1:N], x[i,i] == 0)
@constraint(m, ciclo[i=2:N,j=2:N;i!=j], u[i]-u[j]+N*x[i,j] <= N-1)
#@constraint(m, ciclo[i=1:n,j=1:n;i!=j], u[i]-u[j]+1 <= (n-1)*(1-x[i,j]))
solve(m)
```

```julia
X = getvalue(x)
p = plot_path(pontos, X)
title!(p, "Distância total: $(getobjectivevalue(m))")
p
```

## Prochila do Moblema, digo, Problema da Mochila

O problema da mochila consiste em, dados vários itens e uma limitação, encontrar a melhor seleção de itens que
satisfaz essa limitação.
Tradicionalmente, os itens têm valor e peso, e a limitação é a de ter uma mochila com quantidade máxima de peso.

```julia
type item
    valor :: Int
    peso :: Int
end
```

```julia
N = 20
itens = [item(rand(5:15),rand(15:35)) for i=1:N]
peso_maximo = N * 10
```

```julia
p = plot(leg=false)
x = [0.1; 0.1; 0.9; 0.9; 0.1]
for i = 1:N
    v,w = itens[i].valor, itens[i].peso
    plot!(p, x+i, [0; w; w; 0; 0], c=:red, annotations=(i+0.5,w+1,text("R\$ $v")))
end
xlims!(0.5, N+1.5)
ylims!(0, 3+maximum([itens[i].peso for i = 1:N]))
p
```

O modelo é bastante simples: $x_i = 1$ se o produto $i$ vai na mochila.
Dado um limite $W$ para a mochila, um valor $v_i$ e um peso $w_i$ para cada
produto $i$, temos
$$
\max \sum_{i = 1}^n v_ix_i \\
\mbox{suj. a } \sum_{i=1}^n w_ix_i \leq W \\
x_i \mbox{ Binário}.
$$

```julia
m = Model()
@variable(m, x[1:N], Bin)
@objective(m, Max, sum{x[i]*itens[i].valor, i=1:N})
@constraint(m, sum{x[i]*itens[i].peso, i=1:N} <= peso_maximo)
solve(m)
X = getvalue(x)
getobjectivevalue(m)
```

```julia
p = plot(leg=false)
x = [0.1; 0.1; 0.9; 0.9; 0.1]
for i = 1:N
    v,w = itens[i].valor, itens[i].peso
    c = X[i] == 1.0 ? :blue : :red
    plot!(p, x+i, [0; w; w; 0; 0], c=c, annotations=(i+0.5,w+1,text("R\$ $v")))
end
xlims!(0.5, N+1.5)
ylims!(0, 3+maximum([itens[i].peso for i = 1:N]))
p
```

Existem algumas variações desse problema, como supor que existem mais de um
item com aquelas características. Basta mudarmos $x_i$ de binário para
inteiro para indicar o número de itens daquele tipo levado. Se a quantidade
for limitada, acrescentamos $0 \leq x_i \leq L_i$.

## Problema do Empacotamento (bin packing)

O problema do empacotamento é outro problema inteiro bastante popular.
O problema é parecido com o anterior. Temos $n$ itens com pesos diferentes.
Mas dessa vez, queremos encontrar a menor quantidade de mochilas (iguais)
que são suficientes para carregar tudo.

Dados $n$ produtos com pesos $w_i, i=1,\dots,n$, e mochilas $M_1,M_2,\dots$ com limite de peso $W$
vamos definir $x_{i,j} = 1$ se o item $i$ estiver na mochila $j$.
Note que o número de mochilas não é definido, mas como precisamos limitar o número de mochilas,
podemos usar $n$ mochilas. Note que se algum $w_i > W$, então o problema é infactível.
Caso contrário, podemos usar $n$ mochilas.

Como cada item só pode estar numa mochila, então
$$ \sum_{j=1}^nx_{i,j} = 1. $$

Nosso objetivo é minimizar o número de mochilas. Para isso, vamos criar uma variável $y_j$ que vale 1 se a mochila $j$ for usada.

Nosso objetivo então é
$$ \min \sum_{j=1}^ny_j. $$

Para finalizar, devemos ter os limites da mochila. Se a mochila $j$ não for usada, no entanto,
não podemos ter $x_{i,j} = 1$. Podemos escrever isso da seguinte maneira
$$ \sum_{i=1}^n w_ix_{i,j} \leq W y_j. $$

Se $y_j = 1$, é a restrição da mochila anterior, se $y_j = 0$, a mochila não é usada, e a soma deve ser nula.

O modelo final fica

$$ \min \sum_{j=1}^n y_j \\
\mbox{suj. a }
\sum_{j=1}^n x_{i,j} = 1, \qquad \forall i=1,\dots,n \\
\sum_{i=1}^n w_ix_{i,j} \leq Wy_j, \qquad \forall j=1,\dots,n \\
y_j, x_{i,j} \mbox{ binários}, \qquad \forall 1\leq i,j\leq n.
$$

```julia
n = 30
w = [rand(5:15) for i=1:n]
W = maximum(w)*2
```

```julia
p = plot(leg=false)
x = [0.1; 0.1; 0.9; 0.9; 0.1]
for i = 1:n
    plot!(p, x+i, [0; w[i]; w[i]; 0; 0], c=:red, annotation=(i+0.5,w[i]/2,text(i)))
end
xlims!(0.5, n+1.5)
ylims!(0, 1+maximum(w))
p
```

```julia
m = Model()
@variable(m, x[1:n,1:n], Bin)
@variable(m, y[1:n], Bin)
@objective(m, Min, sum{y[i], i=1:n})
@constraint(m, equ[i=1:n], sum{x[i,j], j=1:n} == 1)
@constraint(m, ineq[j=1:n], sum{w[i]*x[i,j], i=1:n} <= W*y[j])
solve(m)
```

```julia
X = getvalue(x)
```

```julia
Y = getvalue(y)
```

```julia
p = plot(x->W, 0.5, n+1.5, leg=false)
xp = [0.1; 0.1; 0.9; 0.9; 0.1]
h = zeros(Int, n)
for i = 1:n
    j = findfirst(X[i,:] .> 0.99)
    plot!(p, xp+j, h[j] + [0.1; w[i]-0.1; w[i]-0.1; 0.1; 0.1], c=:red, annotation=(j+0.5,h[j]+w[i]/2,text(i)))
    h[j] += w[i]
end
xlims!(0.5, n+1.5)
ylims!(0, 1+W)
p
```

## Um problema mais legal (desafio)

Você tem um conjunto de caixas (paralelepípedos) com dimensões
$a_i \times b_i \times c_i, i = 1,\dots,n$, e containers de dimensões
$L \times A \times P$.

Encontre a disposição dessas caixas de maneira a minimizar o número de
containers utilizados! É permitido guardar as caixas seguindo apenas a
mesma orientação do conteiners, isto é, as caixas podem ser rotacionadas
90$^\circ$ em qualquer dimensão, mas não outros valores.

### Simplifições

- Considere apenas 2 dimensões, e.g., salvar o máximo de imagens retangulares no mínimo de páginas
- Considere apenas a minimização das dimensões, e.g., achar o menor volume do envoltório retangulo.
- Ou ainda, as duas de cima, a menor área retangular contendo todas as imagens.

# Heurísticas

Todos os problemas acima são difíceis. Quando descritos como problemas de
programação linear inteira (mista), normalmente utilizamos métodos como o
Branch-and-Bound ou Branch-and-Cut que envolvem resolver o problema linear
relaxado várias vezes, adicionando restrições ao problema. Isso leva à um
grande problema de tempo, limitando bastante o tamanho do problema possível
de resolver.

Alternativamente, podemos encontrar soluções aproximadas por estratégias
chamadas de heurísticas. As heurísticas usam algumas estratégias específicas
para o problema, e algumas técnicas gerais para melhorar uma aproximação
dada. Existem várias heurísticas, e todas têm o seguinte problema: pode
ser que não funcione, i.e., não ache a solução ótima, mesmo que rodando o
tempo que quisermos.

A vantagem das heurísticas é que elas são, em geral, muito mais baratas de
se calcular, e com frequência encontram soluções boas "o suficiente".
Outro ponto importante é que em geral podemos utilizar processamento paralelo
nas heurísticas com bastante facilidade, o que não acontece muito em PL.

Alguns exemplos de heurística:

- Algoritmo genético: Provavelmente o mais famoso, utilizamos uma população inicial (um conjunto de aproximações) e uma função de qualidade da aproximação. Podemos então escolher os melhores da população e fazer variações neles tentando obter pontos ainda melhores.
- Busca Tabu: Este algoritmo de busca local proíbe movimentos/soluções/vizinhaças/características anteriores por uma quantidade de iterações proibindo o algoritmo de revisitar esses movimentos/sol/...
- Particle Swarm: Não sei, mas tá na moda.
- Simmulated Annealing: Não sei, mas saiu de moda.
